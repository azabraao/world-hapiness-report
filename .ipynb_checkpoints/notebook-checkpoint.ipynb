```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples, adjusted_rand_score

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_csv("student-mat.csv", sep=";")
```

```python
maxGrade = 20;
isApproved = data.G3 >= maxGrade / 2;
data["Situation"] = isApproved
```

```python
data.Situation = data.Situation.map({True: "Approved", False: "Reproved"})
```

```python
data
```

```python
data = data.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
data
```

```python
# Dividindo x e y
# x = data.drop(columns=["Situation","G3"])
x = data.drop(columns=["Situation","G3","G1","G2"])
y = data['Situation']

# Dividindo dados para treino e dados para teste
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

```python
# Treinando o modelo
model = LogisticRegression(verbose=10,solver='lbfgs', max_iter=10000).fit(x_train, y_train)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'pred': y_pred, 'resp': y_test})
```

```python
accuracy_score(y_test, y_pred)
```

```python
_ = plot_confusion_matrix(model, x_test, y_test, cmap = 'Blues', values_format='.0f') #normalize = 'true')
```

```python
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)

_ = plot_confusion_matrix(model, x_test, y_test, cmap = 'Blues', values_format='.0f') #normalize = 'true')
```

```python
plt.rcParams['figure.figsize'] = (12, 7)
plt.scatter(data.higher, data.freetime, alpha=0.5)
plt.xlabel('higher')
plt.ylabel('freetime')
plt.show()
```

```python
kmeans = KMeans(random_state=0).fit(data)
# salva os centroides
centroides = kmeans.cluster_centers_
# salva as labels dos clusters para cada exemplo
y_kmeans = kmeans.predict(data)
```

```python
# plota os dados identificando seus clusters
plt.scatter(data.higher, data.freetime, c=y_kmeans, alpha=0.5, cmap='rainbow')
plt.xlabel('higher')
plt.ylabel('freetime')
# plota os centroides também
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', marker='X', s=200, alpha=0.5)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0) 
kmeans.fit(data)
# salva os centroides
centroides = kmeans.cluster_centers_
# salva as labels dos clusters para cada exemplo
y_kmeans = kmeans.predict(data)

plt.scatter(data.higher, data.freetime, c=y_kmeans, alpha=0.5, cmap='rainbow')
plt.xlabel('higher')
plt.ylabel('freetime')
# plota os centroides também
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', marker='X', s=200, alpha=0.5)
plt.show()
```

```python
# Quantidade de clusters que serão testados
k = list(range(1, 10))

# Armazena das inércias para cada k
inercia = []


# Roda o K-means para cada k fornecido
for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data)
    inercia.append(kmeans.inertia_)

plt.rcParams['figure.figsize'] = (5, 3)
# Plota o gráfico com as inércias
plt.plot(k, inercia, '-o')
plt.xlabel(r'Número de clusters')
plt.ylabel('Inércia')
plt.show()
```
